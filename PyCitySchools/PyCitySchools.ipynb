```python
# Dependencies and Setup
import pandas as pd

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv("Resources/schools_complete.csv")
student_data = pd.read_csv("Resources/students_complete.csv")

# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
students = school_data_complete["Student ID"].count()

schools_gb = school_data_complete.groupby(["school_name", "type"]).sum()
schools_tot = len(schools_gb.index)

schools_gb_mean = school_data_complete.groupby(["school_name"]).mean()
budget = schools_gb_mean["budget"].sum()

math = school_data_complete["math_score"].mean()
reading = school_data_complete["reading_score"].mean()

perc_read = (school_data_complete[school_data_complete.reading_score >= 70]["Student ID"].count())/students * 100
perc_math = (school_data_complete[school_data_complete.math_score >= 70]["Student ID"].count())/students * 100
perc_ov = (school_data_complete[(school_data_complete.reading_score >= 70) & (school_data_complete.math_score >= 70)]["Student ID"].count())/students * 100
```

```python
dist_summary = pd.DataFrame({
    "Total Schools": schools_tot,
    "Total Students": students,
    "Total Budget": "$" + str(budget),
    "Average Math Score": math,
    "Average Reading Score": reading,
    "% Passing Math": perc_math,
    "% Passing Reading": perc_read,
    "% Overall Passing": perc_ov
}, index=[0])

dist_summary
```

```python
school_total_students = school_data_complete.groupby("school_name")["size"].mean()

total_school_budget = school_data_complete.groupby("school_name")["budget"].mean()

reading_m = schools_gb_mean["reading_score"].tolist()
math_m = schools_gb_mean["math_score"].tolist()

reading_students = school_data_complete.query("`reading_score` >= 70")
math_students = school_data_complete.query("`math_score` >= 70")
overall_students = pd.merge(reading_students, math_students, how="inner")

reading_students_t = reading_students.groupby("school_name")["Student ID"].count()
math_students_t = math_students.groupby("school_name")["Student ID"].count()
overall_students_t = overall_students.groupby("school_name")["Student ID"].count()

mulitiindex_df = schools_gb.index.to_frame()
school_type = mulitiindex_df["type"]
```

```python
school_summary = pd.DataFrame({
    "School Type": school_type.to_list(),
    "Total Students": school_total_students.tolist(),
    "Total School Budget": total_school_budget.tolist(),
    "Per Student Budget": (total_school_budget/school_total_students).tolist(),
    "Average Math Score": math_m,
    "Average Reading Score": reading_m,
    "% Passing Math": (math_students_t/school_total_students * 100).tolist(),
    "% Passing Reading": (reading_students_t/school_total_students * 100).tolist(),
    "% Overall Passing": (overall_students_t/school_total_students * 100).tolist()
}, index=[schools_gb_mean.index])

school_summary
```

```python
bot5_df = school_summary.sort_values(by = "% Overall Passing")
top5_df = school_summary.sort_values(by = "% Overall Passing", ascending = False)
```

```python
low = school_summary.loc[school_summary["Per Student Budget"] < 585]
mid1 = school_summary.loc[(school_summary["Per Student Budget"] >= 585) & (school_summary["Per Student Budget"] < 630)]
mid2 = school_summary.loc[(school_summary["Per Student Budget"] >= 630) & (school_summary["Per Student Budget"] < 645)]
high = school_summary.loc[(school_summary["Per Student Budget"] >= 645) & (school_summary["Per Student Budget"] < 680)]

low_read = low["Average Reading Score"].mean()
mid1_read = mid1["Average Reading Score"].mean()
mid2_read = mid2["Average Reading Score"].mean()
high_read = high["Average Reading Score"].mean()

low_math = low["Average Math Score"].mean()
mid1_math = mid1["Average Math Score"].mean()
mid2_math = mid2["Average Math Score"].mean()
high_math = high["Average Math Score"].mean()

low_tot = low["Total Students"].sum()
mid1_tot = mid1["Total Students"].sum()
mid2_tot = mid2["Total Students"].sum()
high_tot = high["Total Students"].sum()

low_read_stu = reading_students.query(("`budget` / `size` < 585"))["Student ID"].count()
mid1_read_stu = reading_students.query(("585 <= (`budget` / `size`) < 630"))["Student ID"].count()
mid2_read_stu = reading_students.query(("630 <= (`budget` / `size`) < 645"))["Student ID"].count()
high_read_stu = reading_students.query(("645 <= (`budget` / `size`) < 680"))["Student ID"].count()

low_math_stu = math_students.query(("`budget` / `size` < 585"))["Student ID"].count()
mid1_math_stu = math_students.query(("585 <= (`budget` / `size`) < 630"))["Student ID"].count()
mid2_math_stu = math_students.query(("630 <= (`budget` / `size`) < 645"))["Student ID"].count()
high_math_stu = math_students.query(("645 <= (`budget` / `size`) < 680"))["Student ID"].count()

low_ov_stu = overall_students.query(("`budget` / `size` < 585"))["Student ID"].count()
mid1_ov_stu = overall_students.query(("585 <= (`budget` / `size`) < 630"))["Student ID"].count()
mid2_ov_stu = overall_students.query(("630 <= (`budget` / `size`) < 645"))["Student ID"].count()
high_ov_stu = overall_students.query(("645 <= (`budget` / `size`) < 680"))["Student ID"].count()
```

```python
school_spending = pd.DataFrame({
    "Average Math Score": [low_math, mid1_math, mid2_math, high_math],
    "Average Reading Score": [low_read, mid1_read, mid2_read, high_read],
    "% Passing Math": [low_math_stu/low_tot *100, mid1_math_stu/mid1_tot *100, mid2_math_stu/mid2_tot *100, high_math_stu/high_tot *100],
    "% Passing Reading": [low_read_stu/low_tot *100, mid1_math_stu/mid1_tot *100, mid2_math_stu/mid2_tot *100, high_math_stu/high_tot *100],
    "% Overall Passing": [low_ov_stu/low_tot *100, mid1_ov_stu/mid1_tot *100, mid2_ov_stu/mid2_tot *100, high_ov_stu/high_tot *100]
}, index = ["<$585", "$585-630", "$630-645", "$645-680"])

school_spending
```
